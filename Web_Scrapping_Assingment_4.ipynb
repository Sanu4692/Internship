Scrape the details of most viewed videos on YouTube from Wikipedia: Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/ You need to find following details: A) Rank B) Name C) Artist D) Upload date E) Views

```python
#1
#importing webdriver
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()  #maximizing window
time.sleep(3)
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"   #"url"
driver.get(url) #loading url
time.sleep(2)

#creating empty Boxex
rank=[]
videon=[]
uploader=[]
views=[]
uploadd=[]

for i in range(1,31):
    #Rank
    dt=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{}]/td[1]".format(i))
    for d in dt:
        rank.append(d.text)
    #video name
    vi=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{}]/td[2]".format(i))
    for c in vi:
        videon.append(c.text)
        videon = [x.split("[")[0] for x in videon] #cleaning data
        
    #Uploader
    l=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{}]/td[3]".format(i))
    for g in l:
        uploader.append(g.text)
    #Views
    vs=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{}]/td[4]".format(i))
    for h in vs:
        views.append(h.text)
   #Upload date
    up=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{}]/td[5]".format(i))
    for c in up:
        uploadd.append(c.text)

time.sleep(2)
#creating dataframe
ans1=pd.DataFrame({})
ans1["Rank"]=rank
ans1["Video Name"]=videon
ans1["Uploader"]=uploader
ans1["Views (billions)"]=views 
ans1["Upload Date"]=uploadd
ans1.set_index("Rank", inplace=True)
ans1
```

2.Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/. You need to find following details: A) Match title (I.e. 1st ODI) B) Series C) Place D) Date E) Time Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#2
#importing webdriver
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window() #maximizing window
time.sleep(3)
url="https://www.bcci.tv/"   #"url"
driver.get(url) #loading url
time.sleep(3)

#clicking on fixture button
fixtures=driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']/div/ul/li/a")
try:
    fixtures.click()

except ElementNotInteractableException:
    driver.get(fixtures.get_attribute('href'))
time.sleep(3)

#creating boxex
mt=[]
series=[]
place=[]
date=[]
ti=[]
month=[]

#Match Title 
d=driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
for i in d:
    mt.append(i.text)

#Series    
s=driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
for i in s:
    series.append(i.text)

#Place
p=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
for i in p:
    place.append(i.text)
    
#Date    
z=driver.find_elements_by_xpath("//span[@class='fixture__date']")
for i in z:
    date.append(i.text)
    
#Month
h=driver.find_elements_by_xpath("//span[@class='fixture__month']")
for i in h:
    month.append(i.text)
    
#time
v=driver.find_elements_by_xpath("//span[@class='fixture__time']")
for i in v:
    ti.append(i.text)
    
time.sleep(2)
#creating dataframe
ans2=pd.DataFrame({})
ans2["Match Title"]=mt
ans2["Series"]=series
ans2["Place"]=place
ans2["Date"]=date 
ans2["Month"]=month
ans2["Date"]=ans2["Date"]+" " + ans2["Month"]   #adding date and month in one column
ans2.drop("Month",axis=1,inplace=True)
ans2["Time"]=ti
ans2
```

Scrape the details of State-wise GDP of India from statisticstime.com. Url = http://statisticstimes.com/ You have to find following details: A) Rank B) State C) GSDP at current price (19-20) D) GSDP at current price (18-19) E) Share(18-19) F) GDP($ billion) Note: - From statisticstimes home page you have to reach to economy page through code.

```python
#4
#importing webdriver
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
time.sleep(3)
url="http://statisticstimes.com/"   #"url"
driver.get(url)  #loading url
time.sleep(2)

#Selecting india in economy dropdown
gdp=driver.find_element_by_xpath("//*[@id='top']/div[2]/div[2]/div/a[3]")

try:
    gdp.click()

except ElementNotInteractableException:
    driver.get(gdp.get_attribute('href'))
time.sleep(3)

#selecting state gdp
driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/ul/li[1]/a").click()

#creating empty boxes
rank=[]
state=[]
gs19=[]
gs18=[]
share=[]
gd=[]

#extraxting data
for i in range(1,34):
    #Rank
    dt=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr[{}]/td[1]".format(i))
    for d in dt:
        rank.append(d.text)
    #State
    vi=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr[{}]/td[2]".format(i))
    for c in vi:
        state.append(c.text)
    #GSDP (Cr INR at Current prices)19-20
    l=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr[{}]/td[3]".format(i))
    for g in l:
        gs19.append(g.text)
    #GSDP (Cr INR at Current prices)18-19
    vs=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr[{}]/td[4]".format(i))
    for h in vs:
        gs18.append(h.text)
   #share
    up=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr[{}]/td[5]".format(i))
    for c in up:
        share.append(c.text)
    #GDP($Billion)    
    dt=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr[{}]/td[5]".format(i))
    for z in dt:
        gd.append(z.text)

time.sleep(2)
#creating dataframe
ans4=pd.DataFrame({})
ans4["Rank"]=rank
ans4["State"]=state
ans4["GSDP (Cr INR at Current prices)19-20"]=gs19
ans4["GSDP (Cr INR at Current prices)18-19"]=gs18 
ans4["Share"]=share
ans4["GDP($Billion)"]=gd
ans4.set_index("Rank", inplace=True) #setting rank column as index
ans4
```

Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ You have to find the following details: A) Dataset name B) Data type C) Task D) Attribute type E) No of instances F) No of attribute G) Year Note: - from the home page you have to go to the Show All Dataset page through code.

```python
#importing webdriver
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()   #maximizing windows
time.sleep(3)
url="https://archive.ics.uci.edu/ml/index.php"   #"url"
driver.get(url)  #loading url
time.sleep(3)

#clicking on view all data sets
driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]").click()


#creating empty boxes
time.sleep(3)
name=[]
dtype=[]
task=[]
at=[]
noi=[]
noa=[]
year=[]


#extracting dataset
for i in range(2,590):
    
#Dataset name
    c=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[1]/table/tbody/tr/td[2]/p".format(i))
    for a in c:
        name.append(a.text)

# Data Type
    z=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[2]/p".format(i))
    for c in z:
        dtype.append(c.text)

#Task
    v=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[3]".format(i))
    for t in v:
        task.append(t.text)

#Attribute type
    y=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[4]".format(i))
    for u in y:
        at.append(u.text)

#NO. of instances
    t=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[5]".format(i))
    for r in t:
        noi.append(r.text)

#No. of attribute
    r=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[6]".format(i))
    for w in r:
        noa.append(w.text)

#Year
    w=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[7]".format(i))
    for y in w:
        year.append(y.text)


    
time.sleep(2)
#creating dataframe
Ans10=pd.DataFrame({})
Ans10["Dataset Name"]=name
Ans10["Data Type"]=dtype
Ans10["Task"]=task
Ans10["Attribute Type"]=at
Ans10["No. of instances"]=noi
Ans10["No. of Attributes"]=noa
Ans10["Year"]=year
Ans10
```
